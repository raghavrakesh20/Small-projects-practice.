```
# Using the 'with' statement as a Context Manager
with open("poem.txt", "w") as f:
    f.write("Digital Humanities explores the intersection of computing and culture.\n")
    f.write("It moves from close reading to distant viewing.")
```

```
# Reading the data
```

```
with open("poem.txt", "r") as f:
    # .read() brings the entire content into a single variable
    content = f.read()
    print(content)
```

```
# : Working with Structured JSON (like author, date, and location) that a simple text file cannot handle.
```

```
import json

# Define a dictionary of cultural metadata
metadata = {
    "title": "Manovich's Cultural Analytics",
    "year": 2020,
    "themes": ["Big Data", "Visual Culture", "Visualization"]
}

# Save this dictionary as a JSON file
with open("data.json", "w") as f:
    json.dump(metadata, f, indent=4)
```

```
# Loading JSON Back into Python
```

```
with open("data.json", "r") as f:
    loaded_data = json.load(f)

# Access a specific part of the data
print(loaded_data["themes"][0])
```
