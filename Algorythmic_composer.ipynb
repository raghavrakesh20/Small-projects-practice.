```python
#importing tools
import music21 as m21
import random
```

```python
# Loading the Training Music
print("Fetching sheet music from the database...")
# Get the file paths for Bach's music
bach_files = m21.corpus.getComposer('bach')

# We will load just the first 5 songs to keep the process fast
training_songs = []
for i in range(5):
    song = m21.converter.parse(bach_files[i])
    training_songs.append(song)

print(f"Successfully loaded {len(training_songs)} songs for the AI to study.")
```

```python
# Extract the Sequence of Notes
print("Extracting notes from the music...")
all_notes = []

# Look at each song one by one
for song in training_songs:
    # Flatten the complex sheet music and extract only the musical events
    for element in song.flat.notes:

        # Check if the event is a single note
        if isinstance(element, m21.note.Note):
            # Add the specific pitch name to our list
            all_notes.append(str(element.pitch))

        # Check if the event is a chord (multiple notes played together)
        elif isinstance(element, m21.chord.Chord):
            # Extract only the highest note of the chord to keep our melody simple
            all_notes.append(str(element.pitches[-1]))

print(f"Extraction complete. The AI has extracted a continuous sequence of {len(all_notes)} notes.")
```

```python
# Building the Markov Chain
print("Building the mathematical brain...")
markov_chain = {}

# We loop through all 1393 notes, stopping just before the very last one
for i in range(len(all_notes) - 1):
    current_note = all_notes[i]
    next_note = all_notes[i + 1]

    # If the AI has not seen this note before, it creates a blank list for it
    if current_note not in markov_chain:
        markov_chain[current_note] = []

    # The AI records what note came next
    markov_chain[current_note].append(next_note)

print(f"Brain built! The AI memorized the patterns for {len(markov_chain)} unique starting notes.")
```

```python
print("Composing new music...")

# Step 1: Pick a starting note that we know exists in our brain
current_note = "C4"
generated_melody = [current_note]

# Step 2: Generate 50 new notes
for i in range(50):
    # Find the list of possible next notes for our current note
    if current_note in markov_chain:
        possible_next_notes = markov_chain[current_note]

        # Randomly pick one note from the probability list
        next_note = random.choice(possible_next_notes)

        # Add it to our new song
        generated_melody.append(next_note)

        # Make the new note the current note for the next loop
        current_note = next_note
    else:
        # If the AI gets stuck, stop the loop
        break

print(f"Composition complete! The AI wrote a melody with {len(generated_melody)} notes.")
print(generated_melody)
```

```python
#Convert and download music
from google.colab import files

print("Converting to a playable MIDI file...")
# Create a blank piece of digital sheet music
new_song_stream = m21.stream.Stream()

# Turn our text notes back into musical objects
for note_name in generated_melody:
    new_note = m21.note.Note(note_name)
    # Make every note the same length (a quarter note)
    new_note.quarterLength = 1.0
    new_song_stream.append(new_note)

# Save the music as a MIDI file
file_name = 'ai_composition.mid'
new_song_stream.write('midi', fp=file_name)

# Trigger the download to your computer
print("Downloading your new AI track...")
files.download(file_name)
```
